### Training the Model 


```python
# Splitting features and target 
x= airline_dataset.drop(columns= 'satisfaction', axis = 1)
y= airline_dataset['satisfaction']
```

```python
print(y)
```

```python
print(x)
```

```python
# Scaling data 
standard_scaler = StandardScaler()
x = standard_scaler.fit_transform(x)
print(x)
```

```python
#Deviidng the data into the training data and test data 

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=25, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#Preparing dataframe to print the models and their respecrive accuracy 
model_accuracy = pd.DataFrame(columns=['Model', 'Accuracy'])
```

### Random Forest

```python
# Create a model using RandomForestClassifier
model = RandomForestClassifier()
```

```python
# Utilize a training set to test model .
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# Utilizing testing set to test model

accuracy = model.score(x_test, y_test)
print('Random Forest Classifier')
print(f'Model accuracy\t\t: {accuracy}')
print(f'Accuracy is Persentage\t: {"{:.1%}".format(accuracy)}')
print(classification_report(y_test, y_pred))
```

```python
# import matplotlib.pyplot as plt
# import seaborn as sns
# from sklearn.metrics import confusion_matrix

# Assuming y_test and y_pred are already defined
con_matrix = confusion_matrix(y_test, y_pred)

# Create a heatmap of the confusion matrix
sns.set(font_scale=1.2)  # Set font size correctly
sns.heatmap(con_matrix, annot=True, annot_kws={'size': 16}, cmap='Blues')  # Correct argument names

plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
plt.title('Confusion Matrix for Random Forest Classifier')

# Show the plot
plt.show()
```

### Logistic Regression

```python
#Create a model using LogisticRegression 
LRmodel = LogisticRegression()
```

```python
# #Utilize training set to test the model 
# acc = LRmodel.score(x_test, y_test)
# print('Logistic Regression scores')
# print(f'Model accuracy\t\t: {acc}')
# print(f'Accuracy in Persentage\t: {"{:1%}".format(acc)}')
# print(classification_report(y_test, y_pred))

# Initialize Logistic Regression model
LRmodel = LogisticRegression()

# Fit the model on the training data
LRmodel.fit(x_train, y_train)

# Now test the model
acc = LRmodel.score(x_test, y_test)
print('Logistic Regression scores')
print(f'Model accuracy\t\t: {acc}')
print(f'Accuracy in Percentage\t: {"{:1%}".format(acc)}')

# Generate predictions and classification report
y_pred = LRmodel.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
#Confusion matrix for logistic regression
con_matrix = confusion_matrix(y_test, y_pred)

#Create a heatmap of the confusion matrix
sns.set(font_scale=1.2) #Set font size
sns.heatmap(con_matrix, annot=True, annot_kws={"size": 16}, cmap='Blues') #Create heatmap 

#Add labels and title 
plt.xlabel('predicted value')
plt.ylabel('exact value')
plt.title('confusion matrix for logistic regression')

plt.show()
```
